```python
import sys

def GenerateGenomePath(k, Text) :
    
    whole_genome = Text[0]
    
    prefix_target = Text[0][: k - 1]
    suffix_target = Text[0][1 :]
    
    for i in range(1, len(Text)) :
        if(suffix_target == Text[i][: k - 1]) :
            suffix_target = Text[i][1 :]
            whole_genome = whole_genome + Text[i][k - 1]
        elif(prefix_target == Text[i][1 :]) :
            prefix_target = Text[i][: k - 1]
            whole_genome = whole_genome + Text[i][0] 
    return whole_genome

#input_file = open("dataset_198_3.txt", 'r')
#Text = input_file.read().splitlines()

Text = sys.stdin.read().splitlines()
k = len(Text[0])

print(GenerateGenomePath(k, Text))
```

```python
import sys

def Overlap(Text, k) :
    
    check = []; dic = {}
    for i in range(0, len(Text)) : check.append([])
    
    for i in range(0, len(Text)) :
        target = Text[i]
        dic[target] = []
        for j in range(0, len(Text)) :
            if(i == j) : continue
            if(target[1:] == Text[j][:k - 1]) : check[i].append(j)
    
    for i in range(0, len(Text)) :
        if(len(check[i]) != 0) :
            for j in range(0, len(check[i])) : dic[Text[i]].append(Text[check[i][j]])
        else : del dic[Text[i]]
                
    return dic.items()

input_file = open("dataset_198_10.txt", 'r')
Text = input_file.read().splitlines()

#Text = sys.stdin.read().splitlines()

print_table = Overlap(Text, len(Text[0]))

for line in print_table :
    printline = "%s -> " % line[0]
    last_node = line[1].pop()
    for node in line[1] : printline += "%s," % node
    printline += "%s" % last_node
    print(printline)
```

```python
import sys

def CompositionK(k, Text) :
    kmers = []
    for i in range(0, len(Text)) : kmers.append(Text[i][:k - 1])
    return list(set(kmers))

def DeBruijnGraphFromEdge(Text, k) :
    
    kmers = CompositionK(k, Text)
    dic = {}
    for kmer in kmers : dic[kmer] = []
    
    for i in range(0, len(Text)) : dic[Text[i][:k - 1]].append(Text[i][1:])

    return dic.items()

#input_file = open("dataset_200_8.txt", 'r')
#Text = input_file.read().splitlines()

Text = input_file.read().splitlines()
k = len(Text[0])

print_table = DeBruijnGraphFromEdge(Text, k)

for line in print_table :
    printline = "%s -> " % line[0]
    last_node = line[1].pop()
    for node in line[1] : printline += "%s," % node
    printline += "%s" % last_node
    print(printline)
```
