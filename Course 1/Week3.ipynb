```python
def ProfileMostProbableKmer(text, k, profile) :

    best_value = 0.0; best_kmer = ""
    
    for i in range(0, len(text) - k + 1) : 
        target = text[i : i + k]
        probability = float(1)
        
        for j in range(0, k) :
            probability *= profile[target[j]][j]
        
        if(probability > best_value) :
            best_kmer = target
            best_value = probability
            
    return best_kmer
```

```python
import math
import numpy as np

def HammingDistance(p, q) :
    count = 0
    for i in range(0, len(p)) :
        if(p[i] != q[i]) : count += 1
    return count

def scoring_motif(motif) : 
    
    dic1 = {'A' : 0, 'C' : 1, 'G' : 2, 'T' : 3}
    dic2 = "ACGT"

    consensus = ""
    
    for i in range(0, len(motif[0])) : 
        
        count_arr = [0.0, 0.0, 0.0, 0.0]
        for j in range(0, len(motif)) : count_arr[dic1[motif[j][i]]] += 1
        index = np.argmax(count_arr)
        consensus += dic2[index]
    
    count = 0
    for i in range(len(motif)) :
        count += HammingDistance(motif[i], consensus)
        
    return count

def generate_matrix(motif_table, length, gene_number) :

    AGTC_table = [[0 for j in range(length)] for i in range(4)]
    
    for i in range(0, length) :

        for j in range(0, gene_number) :

            if(motif_table[j][i] == 'A') : AGTC_table[0][i] += 1
            elif(motif_table[j][i] == 'G') : AGTC_table[1][i] += 1
            elif(motif_table[j][i] == 'T') : AGTC_table[2][i] += 1
            else : AGTC_table[3][i] += 1

        for j in range(0, 4) : AGTC_table[j][i] /= float(AGTC_table[j][i] + 1) / float(gene_number + 4)

    return AGTC_table

def generate_probability(pattern, prob_table) :

    dic = {'A' : 0, 'G' : 1, 'T' : 2, 'C' : 3}
    
    prob = 1.0
    for i in range(0, len(pattern)) : 
        prob *= prob_table[dic[pattern[i]]][i]
    
    return prob

def GreedyMotifSearch(Dna, k, t) :
    
    best_motif = []; best_prob = float(k * t)
    for i in range(0, t) : best_motif.append(Dna[i][0 : k])
        
    for i in range(0, len(Dna[0]) - k + 1) :
        motif = [Dna[0][i : i + k]]
        
        for j in range(1, t) :
            prob = generate_matrix(motif, k, j)
            best_kmer = Dna[j][:k]; best_value = generate_probability(best_kmer, prob)
            
            for l in range(1, len(Dna[j]) - k + 1) :
                target = Dna[j][l : l + k]
                each_prob = generate_probability(target, prob)
                
                if(each_prob > best_value) :
                    best_kmer = target
                    best_value = each_prob
                    
            motif.append(best_kmer)
        
        cal = scoring_motif(motif)
        
        if(cal < best_prob) :
            best_prob = cal
            best_motif = motif
            
    return best_motif

input_file = open("dataset_159_5 (3).txt", 'r')
input_file.readline()

table = input_file.readlines()

GreedyMotifSearch(table, 12, 25)
```
